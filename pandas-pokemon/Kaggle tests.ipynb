# Visualization of Pok√©mon stats

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Load the data

```python
df = pd.read_csv('../data/Pokemon.csv')
df.head()
```

Plot HP versus attack

```python
sns.set(style="ticks", color_codes=True)
sns.jointplot(x="HP", y="Attack", data=df)
```

Attack VS Defense for legendary and non legendary Pokemons

```python
sns.pairplot(df, vars=['Attack','Defense','Total'],  hue='Legendary', diag_kind='kde')
```

```python
sns.violinplot(data=df[['Attack','Defense', 'HP']])
```

```python
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(df.corr(), annot=True, linewidths=.5, ax=ax)
```
